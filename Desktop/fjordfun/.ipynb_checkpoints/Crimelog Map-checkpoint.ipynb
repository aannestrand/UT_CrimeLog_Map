```python
from geopy.geocoders import Nominatim
import re
import pandas as pd

nom = Nominatim(user_agent='my-application')

crimelog = pd.DataFrame(columns=['Address','Latitude','Longitude'])

with open('output.txt', 'r') as f:
    lines = f.readlines()

for line in lines:
    parts = line.split(" ")
    if(re.search("[0-9]", parts[0]) and not(re.search("/", parts[0])) and (len(parts) >= 2) and not(re.search("[a-z]", parts[0]))): 
        line = re.sub("\n","",line)
        line = line + ' AUSTIN'
        crimelog = crimelog.append({'Address': line}, ignore_index=True)
        if(nom.geocode(line) == 'None')
            pass
        else
            crimelog = crimelog.append({'Latitude': (nom.geocode(line)).latitude}, ignore_index=True)
            crimelog = crimelog.append({'Longitude': (nom.geocode(line)).longitude}, ignore_index=True)
    
crimelog
```
